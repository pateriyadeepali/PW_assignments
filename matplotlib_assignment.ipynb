```python
#2
#A scatter chart, also called a scatter plot, is a chart that shows the relationship between two variables. 
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y label")
plt.title("Relatonship btw x and y")
```

```python
#3.
#The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.  
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')
plt.show()
```

```python
# 4.
#Bar graphis a type of data visualization that presents categorical data with rectangular bars. It's used to compare and display the frequency, count, or value of different categories.
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company , profit , color ='#E0B352')
plt.show()
```

```python
#5.
#A box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key statistical measures and provides a visual summary of the spread, central tendency, and potential outliers within the data.
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data, vert=True, labels=['Box 1', 'Box 2'])
plt.xlabel('Boxes')
plt.ylabel('Values')
plt.title('Box Plot ')
plt.show()
```
