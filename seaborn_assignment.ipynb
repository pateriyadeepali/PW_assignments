Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating a variety of informative and attractive statistical graphics. 

1.Scatter Plot-
Scatter plots display individual data points as dots, helping to visualize the distribution and relationship between two numerical variables.
   
2.Line Plot:
Line plots are useful for visualizing trends and changes in data over a continuous interval, often used for time series data or to show the relationship between two variables.

3.Bar Plot:
Bar plots are used to compare categorical data by showing the frequency or distribution of data points within each category. They are great for making comparisons and identifying patterns.

4.Histogram:
Histograms display the distribution of a single numerical variable by dividing it into bins and showing the frequency or count of data points within each bin. They help visualize the data's overall shape and identify patterns.

5.Heatmap:
Heatmaps are used to visualize the relationships between two categorical variables or numerical variables by displaying color-coded values in a matrix format. They are particularly effective for showing correlations and patterns in large datasets.

```python
import matplotlib.pyplot as plt
titanic_data = sns.load_dataset("titanic")

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.boxplot(data=titanic_data, x='pclass', y='age', ax=axes[0])
axes[0].set_title('Box Plot of Age vs. Pclass')
sns.boxplot(data=titanic_data, x='pclass', y='fare', ax=axes[1])
axes[1].set_title('Box Plot of Fare vs. Pclass')
```
